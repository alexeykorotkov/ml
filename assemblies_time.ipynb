```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

#### Подготовка данных

```python
df1 = pd.read_csv('assemblies_assembly_3010.csv', sep=",")
df1
```

```python
df1 = df1.drop(['assembly_box', 'goods_amount', 'status'], axis=1)
df1
```

```python
df1=df1.rename(columns = {'datetime' : 'start'})
df1
```

```python
df2 = pd.read_csv('assemblies_assembled_3010.csv', sep=",")
df2
```

```python
df2 = df2.drop(['assembly_box', 'status'], axis=1)
df2
```

```python
df2=df2.rename(columns = {'datetime' : 'finish'})
df2
```

```python
df3 = pd.merge(df1,df2, how='inner', on='assembly_id', sort=True)
df3
```

```python
df3.info()
```

```python
df3['start'] = pd.to_datetime(df3['start'])
df3['start']
```

```python
df3['finish'] = pd.to_datetime(df3['finish'])
df3['finish']
```

```python
df3['duration'] = df3['finish'] - df3['start']
df3['duration'].sort_values(ascending=False)
df3.head()
```

```python
df3['minutes'] = round(df3['duration'] / pd.Timedelta(minutes=1), 1)
df3
```

```python
df3 = df3[['assembly_id','start','finish','duration','minutes','goods_amount']]
df3
```

#### Визуализация данных

```python
x = [x for x in df3['goods_amount']]
y = [y for y in df3['minutes']]
plt.figure(figsize=(16,8))
plt.title("Скорость сборки заказов", size=16)
plt.xlabel("Кол-во товаров в заказе, шт.")
plt.ylabel("Время сборки заказа в магазине, мин.")
plt.scatter(x,y)
```

#### Обучение модели linearRegression

```python
model = np.polyfit(x, y, 1)
model
```

```python
predict = np.poly1d(model)
predict
```

```python
goods_amount = 40
predict(goods_amount)
```

```python
from sklearn.metrics import r2_score
r2_score(y, predict(x))
```

```python
x_lin_reg = range(0, 61)
y_lin_reg = predict(x_lin_reg)
plt.figure(figsize=(16,8))
plt.scatter(x, y)
plt.plot(x_lin_reg, y_lin_reg, c = 'r')
```
